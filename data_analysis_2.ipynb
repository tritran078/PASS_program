## **Mean**: 
- mean is the average value of the data.
- to get mean, one must sum the total value of the same type and divided by the total number of values.
- one can only find the mean value of the data if the data type is numerical.
taken for exmaple in the datasets "gifted.csv"
- Ex: the average IQ score of a child int he datasets is the sum of column "score" divided by the number of observation in that row.

### finding the mean manually:

```python
iq_column = df.iloc[:,0] # get only the first row data (score)
sum  = iq_column.sum()   # get the sum of the iq column
mean = sum / len(iq_column)  # divide the sum by total length

print(f"The average IQ of the children is {mean:.2f}") 
```

### or we can find the mean using numpy package

```python
iq_column = df.iloc[:,0]
mean = np.mean(iq_column)

print(f"The average IQ of the children is {mean:.2f}") 
```

#  

## **Median:**
- the middle value for an ordered numerical dataset.
- if 1,4,8,67,89 is a datasets, the median is 8
- if the datasets is even: the median is the average value between the middle+1 and middle-1 position.


### using the Father's IQ as an example for median, we can first sort the data in acsending order, then fidn the middle value and thats our median.


```python
f_iq = df.iloc[:,1]                          # the column for father's IQ

sorted_f_iq = sorted(f_iq)                   # sort the data in ascending order
lower_mid = sorted_f_iq[len(f_iq) // 2 - 1]  # get the lower value in mid 
upper_mid = sorted_f_iq[len(f_iq) // 2]      # get the upper value in mid 
median = (lower_mid + upper_mid) / 2         # find the middle values

print(f"The median value for Father's IQ is {median:.2f}")
```

### or we can use the .median() function to find the median.

```python
f_iq = df.iloc[:,1]                      # the column for father's IQ
median = f_iq.median()

print(f"The median value for Father's IQ is {median:.2f}")
```

## **Mode:**
- mode is the most frequently occured values in the dataset
- for example, in a list of values: 1,2,2,3,4,5,2,2,7,2. 2 is the mode

### we can find mode in "speak" using the function .mode() since finding it manually is a little tedious.

```python
speak = df.iloc[:,4]
mode_ = speak.mode()

print(f"The most common month in which children started speaking is month {mode_[0]}")
```

## **Variance:**

- Variance measures how spread out the dataset is
- If the data is focused around one value, the variance is small
- If the data is spread out and occupied a wide area, the variance is big
- We can find the variance by finding the mean, subtracing the mean with each of the value, square each of the diffrence (to avoid negative values cancelling out), then average all the values


### The example below shows the variance of Mother's IQ

```python
m_iq = df.iloc[:,2] # getting the values for Mother's IQ
mean = m_iq.mean()   # getting the mean value
diff_m_iq = m_iq - mean # subtract each value in mother iq to mean
diff_squared = diff_m_iq ** 2  # squared each value to avoid negative values canceling

variance = diff_squared.mean()  # averaging the different squared.

print(f"The Variance of the Mother's IQ is {variance:.2f}")
```

### or we can use the fucntion var() in the numpy package

```python
m_iq = df.iloc[:,2] # getting the values for Mother's IQ

variance = np.var(m_iq)

print(f"The Variance of the Mother's IQ is {variance:.2f}")
```

# 


## **Standard deviation:**

- eventhough the variance measure the spread out of a data sets, they uses the squared values difference.
- to know what the average deviatio between each value from the mean, we just need to square root the Variance.
- if the Variance for Mother's IQ is 41.14 the actual deviation between each Mother's IQ from the mean is √41.14

### we can find the Standard Deviation by just squaring the Variance

```python
std_dev = np.sqrt(variance)   # takign the square root of the Variance

print(f"The standard deviation of each Mother's IQ is {std_dev:.2f}")
```

### or we can find the Standard Deviation using .std()

```python
standard_dev = m_iq.std(ddof = 0) # ddof = 0 so the numpy does not take data by sample 

print(f"The standard deviation of each Mother's IQ is {standard_dev:.2f}")
```

## **Uniform Distribution:**

- uniform distribution is  a type of distribution where the data sets value are spread equally.
- all the range has equal amount of values. making almost like a rectangle shape graph
- we can think of uniform distribution means "distributing equally" across each columns.


```python
uniform_dist = np.random.uniform(low = 0,high = 100, size = 1000)

fig = px.histogram(uniform_dist, nbins= 10000)
fig.show()
```

### from the graph above in which 1000 values is generated ranging from 0 to 100, each value appear roughly one time at each bin. distributing uniformly withotu clustering into a specific value range.

# Normal Distribution: 
- normal distribution is seen everywhere. from average household income, average IQ, SAT/ACT score to human height are normaly distributed
- The distribution has the shape of a bell curve, where values are mostly clustered into the middle. the specific shape depends on what the data, or the shape can be the Standard Normal Distribution

### here we can the IQ in children distribution and see the frequency value graph

```python
iq = df.iloc[:,0]

fig = px.histogram(iq, nbins = 40)

fig.show()
```

### The data show signs of clustering into the middle, but with only 36 entries, the bell shape is not fully manifested, a normal distribution is best shown with 500+ samples.

### Variance: how spread out the data is
