# ***STANDARD DEVIATION, VARIANCE ACTIVITY***

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import numpy as np 
pio.renderers.default = "iframe"
import requests
from io import StringIO
url = "https://www.openintro.org/data/csv/epa2012.csv"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
response = requests.get(url, headers=headers)
response.raise_for_status() # Raise an exception for bad status codes
df = pd.read_csv(StringIO(response.text))
```

```python
df
```

### - one numerical column of choice is the combined MPG of vehicles.

```python
mpg = df.iloc[:, 11]
```

```python
print(mpg.describe())
print(mpg.median())
```

# what can we infer from the mean ?

### based on the statistic of the data we can see that the mean of the data is around 21.9 miles per gallon. this means that on average every car has around 21.9 miles per gallon. meanign that if we were to pick a car at random from the datasets, we would likely to get a MPG value close to 21.9

### the median tells us that the mean is slightly higher then median, implying that there are cars that has MPG way over the average range, which slightly pulls the mean up.

```python
fig = px.histogram(mpg)
fig.show()
```

### from the plot we can see that the data is skewed towards the left, trailing off further on the right. mos tof the values MPG are focuses near the 0-40 range, with some that has exceptional MPG up to the 100s.

```python
fig = px.box(mpg)
fig.update_layout(width = 300, height =500, title = "Combined MPG", yaxis_title="MPG Range")
fig.update_yaxes(range=[0,60])
fig.show()
```

### from the box plot, we can see the upper fence range as 35 while the lower fence range as 10, meaning the majority number of car's MPG lies within the 35-10 range, any cars that has MPG outside these values are considered outliers. we can also infer the interquartile range, in which the majority half lies within 25 to 18 MPG. 

# ***PRACTICE QUESTIONS***

## why can't we simple average deviation from the mean ?
- we can't simply average deviation from the mean because there would be positives and negatives deviation values, in which the sum total would be zeros. so we square them to avoid negative values. we uses square instead of taking the absolute values to be able to simplifies the math and make the curve of those values differentiable.
## why are deviations squared ?
- deviations are squared to avoid negatives values, which if we add normaly would sumup to zero.
## why do we divide by n-1 for a sample ?
- because sample mean are derived biasly from the sample, in which its deviation is smaller then the actual population mean. the mean of the sample is not an independent external reference, and it must follow the constraint in which the values up until n-1 are free while the last value is locked in to make the sum equals zero, losing one degree of freedom. therefore we remove the last value to remove the bias.
## can SD ever be  negative ?
- no, SD can not be negative because it is the square root of the squared values of the deviations. so it is mathematically not possible to be negative.

## which dataset is more consistent: SD = 2 or SD = 15 ?
- SD = 2 is more consistent because it means the data is less spread out, in which on average the data values are only spread out by + - 2 instead of + - 15.

## how do outliers affect variance and standard deviation ?
- Standard Deviation are heavily sensitive to outliers, in which a few outliers can majorly influence the data.

## when is standard deviation more useful than variance ?
- standard deviation are usually always more useful then variance because they are in the original unit of the data beign measured, so its easier to compare and perhaps visually view the datasets given standard deviation.
