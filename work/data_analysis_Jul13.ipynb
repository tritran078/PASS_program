# ***Linear Regression Data Analysis***

```python
import statsmodels.api as sm
import pandas as pd
import plotly.express as px
import plotly.io as pio
import numpy as np 
pio.renderers.default = "iframe"
import requests
from io import StringIO
url = "https://www.openintro.org/data/csv/fastfood.csv"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
response = requests.get(url, headers=headers)
response.raise_for_status() # Raise an exception for bad status codes
df = pd.read_csv(StringIO(response.text))
```

```python
df.head()
```

## identify the relationship between calorie and cholesterol in fast food, will cholesterol also increase for food with more calorie.

first we check whether the data sets has any NaN values.

```python
print(df.isna().sum())

print(df.info())
```

vitamin A, C and calcium is missing a lot of values so we will not be assesing those.

```python
fig = px.scatter(df, x = "calories", y = "cholesterol", title = "Calorie vs cholesterol",
                labels={"calories": "Calories (kcal)", "cholesterol": "Cholesterol (mg)"},
                color = "restaurant")
fig.show()
```

we can see a faintly positively correlated relationship between calories and cholesterol. proving that food with high calorie in fast food restaurant often asscociated with high cholesterol.

in order to fully see the relationship between the calorie and the cholesterol, we can use the linear regression line to map out a linear relationship between the two variable. 

we will use the function polyfit to get the variable for the slope and the y intercept of the line. 

```python
x = df["calories"]
y = df["cholesterol"]
m , b = np.polyfit(x,y,1)

print("slope:", m)
print("y-intercept:", b)


df["y_pred_cholesterol"] = m * df["calories"] + b

fig = px.scatter(df, x = "calories", y = "cholesterol",
                 labels={"calories": "Calories (kcal)", "sodium": "Sodium (mg)"},
                 color = "restaurant")

fig.add_scatter( x = df["calories"], y = df["y_pred_cholesterol"], name = "Regression prediction line", line=dict(color="red"))

fig.show()

```

## we can see the line following the trend of each the point, mapping the relationship between calorie and cholesterol in fast food Linearly

the y intercept value b is the the mg of cholesterol if the calorie in the food were to be zero. However, the value of B is negative, therefore we can say that a fastfood item can have some calorie and zero cholesterol and not the other way around.

the slope tells how much the cholesterol is with respect to calorie, in which this value is 0.17. therefore, for every 1 calorie being added to the food, about 0.17 mg of cholesterol is also added.

# OSL METHOD

```python
X = sm.add_constant(df["calories"])  # adds intercept
model = sm.OLS(df["cholesterol"], X).fit()
print(model.summary())
```

- we can see the coef value as the y intercept and the 0.1705 as the slope

- the "std err" measures how much the values "wooble" around the line, meaning that ,most of the true value is + - 3.841 in the y direction and 0.006 in the x direction.
